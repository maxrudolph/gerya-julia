# Opening HDF5 File

```julia
using HDF5
using PyPlot

fname = "SeaFreeze_table_with_density.hdf5"
# Displaying all of the file's information
fid = h5open(fname, "r")
```

```julia
# Reading Data from file
entropy = fid["Entropy"];
pressure = fid["Pressure"];
temperature = fid["Temperature"];
density = fid["Density"];
```

```julia
# Displaying the contents in Entropy group 
for obj in entropy
  data = read(obj)
  display(data)
end

# Accessing Entropy contents that I want
ed = read(entropy, "Data");
```

```julia
# Displaying the data in Pressure group 
for obj in pressure
  data = read(obj)
  display(data)
end

# Accessing Pressure content that I want
pd = read(pressure, "Data");
```

```julia
# Displaying the data in Temperature group 
for obj in temperature
  data = read(obj)
  display(data)
end

# Accessing Temperature content that I want
td = read(temperature, "Data");
```

```julia
# Displaying the data in Density group 
for obj in density
  data = read(obj)
  display(data)
end

# Accessing Density content that I want
rhod = read(density, "Data");
```

# Closing HDF5 File

```julia
# Close file
close(fid)
```

```julia
using Interpolations
```

```julia
entropy = ed;
density = rhod;
pressure = pd*1e6;
temperature = td;

# linear interpolation for entropy
interp_linear1 = linear_interpolation((pressure,temperature),entropy)

output1 = zeros(size(entropy))
    
for i=1:size(temperature,1)
    for j=1:size(pressure,1)
        output1[j,i]=interp_linear1(pressure[j],temperature[i])
    end
end


# linear interpolation for density
interp_linear2 = linear_interpolation((pressure,temperature),density)

output2 = zeros(size(density))
    
for i=1:size(temperature,1)
    for j=1:size(pressure,1)
        output2[j,i]=interp_linear2(pressure[j],temperature[i])
    end
end
```

```julia
figure()
pcolor(temperature,pressure,output1-ed,cmap=get_cmap("viridis"))
title(L"Linear\,\,Interpolation\,\,for\,\,Entropy")
colorbar(label=L"Entropy\,(J/K/kg)")
gca().set_xlabel(L"Temperature\,(K)")
gca().set_ylabel(L"Pressure\,(Pa)")
gca().spines["top"].set_visible(false) # Hide the top edge of the axis
gca().spines["right"].set_visible(false) # Hide the right edge of the axis
gca().xaxis.set_ticks_position("bottom")
gca().yaxis.set_ticks_position("left")
gca().spines["left"].set_position(("axes",-0.03)) # Offset the left scale from the axis
gca().spines["bottom"].set_position(("axes",-0.03)) # Offset the bottom scale from the axis
show()

figure()
pcolor(temperature,pressure,output2-rhod,cmap=get_cmap("viridis"))
title(L"Linear\,\,Interpolation\,\,for\,\,Density")
colorbar(label=L"Density\,(kg/m^{3})")
gca().set_xlabel(L"Temperature\,(K)")
gca().set_ylabel(L"Pressure\,(Pa)")
gca().spines["top"].set_visible(false) # Hide the top edge of the axis
gca().spines["right"].set_visible(false) # Hide the right edge of the axis
gca().xaxis.set_ticks_position("bottom")
gca().yaxis.set_ticks_position("left")
gca().spines["left"].set_position(("axes",-0.03)) # Offset the left scale from the axis
gca().spines["bottom"].set_position(("axes",-0.03)) # Offset the bottom scale from the axis
show()
show()
```
